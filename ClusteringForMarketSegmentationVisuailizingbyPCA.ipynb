```
mall_customers = pd.read_csv("/gdrive/My Drive/Mall_Customers.csv")
```

Lets look at the Original Dataset ( displaying top 5 Rows )

```
mall_customers.head()
```

```
X = mall_customers.to_numpy()
```

```
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans

from sklearn.decomposition import PCA
```

Will use only last 3 Features i.e Age, Annual Income and Spending Score

## Apply PCA to X to transform it into 2 feature Dataset

We will plot the clustering classes using PCA output matrix, so visualize the clear seperation among different classes

```
pca = PCA(2)
X_pca = pca.fit_transform(X[:, 2:])
```

```
X_pca.shape
```

Applying KMeans Clustering

**Number of Classes = 3**

```
y_pred_n3 = KMeans(n_clusters=3).fit_predict(X[:, 2:])
```

```
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X_pca[:, 0],X_pca[:, 1], c=y_pred_n3)
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.title("Clustering with k = 3")
print("\n")
```

**Number of Classes = 4**

```
y_pred_n4 = KMeans(n_clusters=4).fit_predict(X[:, 2:])
```

```
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X_pca[:, 0],X_pca[:, 1], c=y_pred_n4)
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.title("Clustering with k = 4")
print("\n")
```

## Applying MininBatchKMeans 

**Number of Clusters = 3**

```
y_pred_mk_n3 = MiniBatchKMeans(n_clusters=3).fit_predict(X[:,2:])
```

```
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X_pca[:, 0],X_pca[:, 1], c=y_pred_mk_n3)
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.title("Clustering with k = 3")
print("\n")
```

**Number of Clusters = 4**

```
y_pred_mk_n4 = MiniBatchKMeans(n_clusters=4).fit_predict(X[:,2:])
```

```
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.scatter(X_pca[:, 0],X_pca[:, 1], c=y_pred_mk_n4)
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.title("Clustering with k = 4")
print("\n")
```
